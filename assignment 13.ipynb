```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32

    def to_kelvin(self):
        return self.celsius + 273.15


# Driver Code
temp = Temperature(25)

print("Temperature in Celsius:", temp.celsius)
print("Temperature in Fahrenheit:", temp.to_fahrenheit())
print("Temperature in Kelvin:", temp.to_kelvin())
```

```python
class Interest:
    def __init__(self, principal, rate, time):
        self.principal = principal
        self.rate = rate
        self.time = time

    def simple_interest(self):
        si = (self.principal * self.rate * self.time) / 100
        return si

    def compound_interest(self):
        amount = self.principal * (1 + self.rate / 100) ** self.time
        ci = amount - self.principal
        return ci


# Driver Code
obj = Interest(10000, 5, 2)

print("Simple Interest =", obj.simple_interest())
print("Compound Interest =", round(obj.compound_interest(), 2))
```

```python
class CoffeeMachine:
    def __init__(self, water, coffee, milk):
        self.water = water
        self.coffee = coffee
        self.milk = milk

    def make_latte(self):
        if self.water >= 200 and self.coffee >= 20 and self.milk >= 150:
            self.water -= 200
            self.coffee -= 20
            self.milk -= 150

            print("Latte is ready!")
            print("Remaining Water:", self.water, "ml")
            print("Remaining Coffee:", self.coffee, "g")
            print("Remaining Milk:", self.milk, "ml")
        else:
            print("Not enough resources to make Latte.")


# Driver Code
machine = CoffeeMachine(300, 100, 200)

machine.make_latte()
```

```python
class Vehicle:
    def __init__(self, name, max_speed):
        self.name = name
        self.max_speed = max_speed

    def display(self):
        print("Vehicle Name:", self.name)
        print("Maximum Speed:", self.max_speed, "km/h")


class Bus(Vehicle):
    def __init__(self, name, max_speed, seating_capacity):
        super().__init__(name, max_speed)
        self.seating_capacity = seating_capacity

    def reservation(self):
        print("Reservation Available")
        print("Seating Capacity:", self.seating_capacity)

    def display(self):
        super().display()
        print("Seating Capacity:", self.seating_capacity)


# Driver Code
bus = Bus("Volvo", 120, 50)

bus.display()
bus.reservation()
```
