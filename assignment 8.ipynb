```python
from tkinter import *
import csv
import os

# Function to save data
def save_contact():
    name = name_entry.get()
    email = email_entry.get()
    mobile = mobile_entry.get()

    if name and email and mobile:
        file_exists = os.path.isfile("address_book.csv")

        with open("address_book.csv", "a", newline="") as file:
            writer = csv.writer(file)

            # Write header if file is new
            if not file_exists:
                writer.writerow(["Name", "Email", "Mobile"])

            writer.writerow([name, email, mobile])

        status_label.config(text="Contact Saved Successfully!")

        # Clear entries
        name_entry.delete(0, END)
        email_entry.delete(0, END)
        mobile_entry.delete(0, END)

    else:
        status_label.config(text="Please fill all fields!")

# Create Window
root = Tk()
root.title("Address Book")
root.geometry("350x250")

# Labels
Label(root, text="Name").grid(row=0, column=0, padx=10, pady=10)
Label(root, text="Email").grid(row=1, column=0, padx=10, pady=10)
Label(root, text="Mobile").grid(row=2, column=0, padx=10, pady=10)

# Entry Boxes
name_entry = Entry(root, width=30)
name_entry.grid(row=0, column=1)

email_entry = Entry(root, width=30)
email_entry.grid(row=1, column=1)

mobile_entry = Entry(root, width=30)
mobile_entry.grid(row=2, column=1)

# Button
Button(root, text="Save Contact", command=save_contact).grid(
    row=3, column=1, pady=15
)

# Status Label
status_label = Label(root, text="")
status_label.grid(row=4, column=1)

root.mainloop()
```
