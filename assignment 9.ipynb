```python
# Module to find Kaprekar Constant (6174)

def kaprekar(num):
    count = 0

    while num != 6174:
        num_str = str(num).zfill(4)

        # Arrange digits
        asc = int("".join(sorted(num_str)))
        desc = int("".join(sorted(num_str, reverse=True)))

        result = desc - asc

        print(f"Step {count + 1}: {desc:04d} - {asc:04d} = {result:04d}")

        num = result
        count += 1

        if num == 0:
            print("Invalid number! All digits are same.")
            break

    if num == 6174:
        print("\nKaprekar Constant (6174) reached!")

# User Input
number = int(input("Enter a 4-digit number: "))
kaprekar(number)
```
