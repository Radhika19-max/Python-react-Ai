```python
t1 = (1, 2)
t2 = (3, 4)
t3 = (5, 6)

new_tuple = t1 + t2 + t3

print("Joined Tuple:", new_tuple)
```

```python
t = (10, 20, 30, 40, 50)

element = 30

if element in t:
    print(element, "exists in the tuple")
else:
    print(element, "does not exist inthe tuple")
```

```python
numbers = (12, 45, 7, 23, 89, 15)

total = 0
highest = numbers[0]
lowest = numbers[0]

for num in numbers:
    total += num
    
    if num > highest:
        highest = num

    if num < lowest:
        lowest = num

print("Total =", total)
print("Highest =", highest)
print("Lowest =", lowest)
```

```python
n = (3, 14, 7, 22, 9, 41, 18, 5)

filtered = ()

for num in n:
    if num > 10:
        filtered += (num,)

print("Filtered Tuple:", filtered)
```

```python
s = {"cat", "dog", "bird", "fish"}

count = 0

for item in s:
    count += 1

print("Number of elements:", count)
```

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}

combined = s1 | s2

print("combined set:", combined)
```

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

common = s1 & s2

print("common Elements:", common)
```
