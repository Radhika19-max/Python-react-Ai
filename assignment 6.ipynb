```python
def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
    
print(maximum(10, 25, 15))
```

```python
def distinct_list(first):
   result = []
   for item in first:
      if item not in result:
         result.append(item)
   return result

print(distinct_list([1, 2, 3, 4, 4, 5]))
```

```python
def multiply_list(first):
    result = 1
    for num in first:
        result *= num
    return result

print(multiply_list([2, 3, 4]))
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *=i
    return fact

print(factorial(5))
```

```python
def reverse_string(text):
    return text[::-1]

print(reverse_string("Python"))
```

```python
def check_range(num):
    if 10 <= num <= 50:
        print(num, "is written the range")
    else:
        print(num, "is outside the range")

check_range(25)
```

```python
def print_even(first):
    for num in first:
        if num % 2 == 0:
            print(num)

print_even([1, 2, 3, 4, 5, 6, 7, 8])
```

```python
def is_prime(n):
    if n <= 1:
        return False
    
    for i in range(2, n):
        if n % i == 0:
            return False
        
    return True

num = 13

if is_prime(num):
    print(num,"is Prime")
else:
    print(num," is Not Prime")
```

```python
def count_case(text):
    upper = 0
    lower = 0
    
    for ch in text:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower +=1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

count_case ("Hello World Letter")
```
