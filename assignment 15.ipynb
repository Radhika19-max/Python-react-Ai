```python
import requests

try:
    stock = input("Enter Stock Symbol (Example: AAPL, MSFT, TSLA): ").upper()

    url = f"https://query1.finance.yahoo.com/v8/finance/chart/{stock}"

    response = requests.get(url)
    response.raise_for_status()

    data = response.json()

    result = data["chart"]["result"][0]["meta"]

    print("\n------ LIVE STOCK DATA ------")
    print("Company Symbol :", result["symbol"])
    print("Current Price  :", result["regularMarketPrice"])
    print("Previous Close :", result["previousClose"])
    print("Currency       :", result["currency"])
    print("Exchange       :", result["exchangeName"])

except requests.exceptions.ConnectionError:
    print("No Internet Connection.")

except requests.exceptions.HTTPError:
    print("HTTP Error Occurred.")

except KeyError:
    print("Invalid Stock Symbol.")

except Exception as e:
    print("Error:", e)
```

```python
class Rectangle:

    def __init__(self, length, width):
        if length <= 0 or width <= 0:
            raise ValueError("Length and Width must be greater than zero.")

        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

try:
    l = float(input("Enter Length: "))
    w = float(input("Enter Width: "))

    r = Rectangle(l, w)

    print("Area =", r.area())
    print("Perimeter =", r.perimeter())

except ValueError as e:
    print(e)

except Exception as e:
    print("Error:", e)
```

```python
import math

class Circle:

    def __init__(self, radius):
        if radius <= 0:
            raise ValueError("Radius must be greater than zero.")

        self.radius = radius

    def diameter(self):
        return 2 * self.radius

    def area(self):
        return math.pi * self.radius ** 2

    def circumference(self):
        return 2 * math.pi * self.radius

try:
    r = float(input("Enter Radius: "))

    c = Circle(r)

    print("Diameter =", c.diameter())
    print("Area =", c.area())
    print("Circumference =", c.circumference())

except ValueError as e:
    print(e)

except Exception as e:
    print("Error:", e)
```

```python
class Temperature:

    def __init__(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32

    def to_kelvin(self):
        return self.celsius + 273.15

try:
    c = float(input("Enter Temperature in Celsius: "))

    t = Temperature(c)

    print("Fahrenheit =", t.to_fahrenheit())
    print("Kelvin =", t.to_kelvin())

except ValueError:
    print("Please enter a valid number.")

except Exception as e:
    print("Error:", e)
```

```python
class Interest:

    def __init__(self, p, r, t):
        if p < 0 or r < 0 or t < 0:
            raise ValueError("Values cannot be negative.")

        self.p = p
        self.r = r
        self.t = t

    def simple_interest(self):
        return (self.p * self.r * self.t) / 100

    def compound_interest(self):
        return self.p * ((1 + self.r/100) ** self.t) - self.p

try:
    p = float(input("Principal: "))
    r = float(input("Rate: "))
    t = float(input("Time: "))

    obj = Interest(p, r, t)

    print("Simple Interest =", obj.simple_interest())
    print("Compound Interest =", obj.compound_interest())

except ValueError as e:
    print(e)

except Exception as e:
    print("Error:", e)
```

```python
class CoffeeMachine:

    def __init__(self, water, coffee, milk):
        if water < 0 or coffee < 0 or milk < 0:
            raise ValueError("Resources cannot be negative.")

        self.water = water
        self.coffee = coffee
        self.milk = milk

    def status(self):
        print("\nResources Available")
        print("Water :", self.water, "ml")
        print("Coffee:", self.coffee, "g")
        print("Milk  :", self.milk, "ml")

try:
    water = int(input("Water (ml): "))
    coffee = int(input("Coffee (g): "))
    milk = int(input("Milk (ml): "))

    machine = CoffeeMachine(water, coffee, milk)
    machine.status()

except ValueError as e:
    print(e)

except Exception as e:
    print("Error:", e)
```
